# Snowpark Container Services -- Scenario: "Services Functions"

Services Function allow you to call your data processing service as part of a regular Snowflake SQL query.

In the event you have a specific data processing operation that doesnt fit into the Snowflake Warehouse, you can write your own custom code to do the transformation you need.

We will leverage this capability to deploy our fine-tuned NLP model onto GPU infrastructure, and generate embeddings.
